```python
import random
import time

wizard = {'strength': 15,'defence': 30,'magic': 60,'health': 80,'gold':150}
dwarf = {'strength': 40,'defence': 40,'magic': 10,'health': 110,'gold':200}
elf = {'strength': 25,'defence': 35,'magic': 25,'health': 120,'gold':100}

class Character:
    
    def __init__(self, strength, defence, magic, health, gold):
        self.strength = strength
        self.defence = defence
        self.magic = magic
        self.health = health
        self.gold = gold
        
    def show_info(self):
        print ("Your strength: {}, defence: {}, magic: {}, gold: {} and health: {}".format(self.strength, self.defence, self.magic, self.gold, self.health))


class Enemy:
    def __init__(self, strength, defence, health):
        self.strength = strength
        self.defence = defence
        self.health = health
```

```python
def attack(attacker_strength, defender_defence):
    if defender_defence > attacker_strength:
        damage = defender_defence - attacker_strength
    else:
        damage = attacker_strength - defender_defence
    return damage
```

```python
while True:
    print ("Choose one of the following options: ")
    print ("1. Play Game")
    print ("2. Exit")
    play = int(input("Enter your choice: "))
    if play == 1:
        skeleton = Enemy(20,20,30)
        scarycat = Enemy(100,10,10)
        flag = True

        choice = input("Select one of the characters: Dwarf , Elf or Wizard? ").lower()
        if choice == 'wizard':
            user = Character(wizard['strength'],wizard['defence'],wizard['magic'],wizard['health'],wizard['gold'])
        elif choice == 'elf':
            user = Character(elf['strength'],elf['defence'],elf['magic'],elf['health'],elf['gold'])
        elif choice == 'dwarf':
            user = Character(dwarf['strength'],dwarf['defence'],dwarf['magic'],dwarf['health'],dwarf['gold'])
        else:
            print ("Invalid choice")
            
        print ("Since you have created a new character, you get 5 bonus points.")
        print ("Your current stats are:")
        user.show_info()
        print ("In which component do you want to add these 5 points? You can write s for strength, h for health, m for magic, d for defence and g for gold.")
        comp = input("Enter your choice: ").lower()
        if comp == 'h':
            user.health += 5
        elif comp == 's':
            user.strength += 5
        elif comp == 's':
            user.strength += 5
        elif comp == 's':
            user.strength += 5
        elif comp == 's':
            user.strength += 5
        else:
            print ("Inv")

        print ("Scenario 1: ")
        print ("There are two roads in front of you. One of them is very beautiful and full of greenery. The other one is weary.")
        print ("You have to choose a road. Type b for choosing the beautiful one and w for weary.")
        road = input("Enter your choice: ").lower()
        if road == 'b':
            while True:
                #User strikes
                damage_user = attack(user.strength,skeleton.defence)
                skeleton.health -= damage_user
                print ("The skeleton health is {}".format(skeleton.health))
                if skeleton.health <= 0:
                    print ("Congrats! You defeated the skeleton.")
                    print ("You can go forward.")
                    break
                    time.sleep(1)
                #Skeleton strikes
                print ("It's the skeleton's turn now!")
                time.sleep(1)
                damage_skeleton = attack(skeleton.strength, user.defence)
                user.health -= damage_skeleton
                if user.health <= 0:
                    print ("Game Over! You lose!")
                    flag = 0
                    break  
        else:
            user.gold += 10
            user.health += 10
            print ("Hello! You have earned some money and health.")
            n = input("Do you want to see you statistics? Say y for yes and n for no: ").lower()
            if n == 'y':
                user.show_info()

        if flag == 1: 
            print ("Scenario 2: ")
            print ("You have two fruits in front of you. An apple and a banana. Choose one of the fruits:")
            print ("Type a for apple and b for banana")
            fruit = input("Enter your choice: ").lower()
            if fruit == 'a':
                user.gold += 10
                user.health += 10
                print ("Hello! You have earned some money and health.")
                n = input("Do you want to see you statistics? Say y for yes and n for no: ").lower()
                if n == 'y':
                    user.show_info()
            else:
                ch = random.randint(0,1)
                if ch == 0:
                    print ("No harm No gain!")
                    print ("You can go forward.")
                else:
                    while True:
                    #User strikes
                        damage_user = attack(user.strength,scarycat.defence)
                        scarycat.health -= damage_user
                        print ("The scary cat's health is {}".format(scarycat.health))
                        if scarycat.health <= 0:
                            print ("Congrats! You defeated the scarycat.")
                            print ("You can go forward.")
                            break
                        time.sleep(1)
                        #Scary cat strikes
                        print ("It's the scary cat's turn now!")
                        time.sleep(1)
                        damage_cat = attack(scarycat.strength, user.defence)
                        user.health -= damage_cat
                        if user.health <= 0:
                            print ("Game Over! You lose!")
                            flag = 0
                            break 

            if flag == 1:
                print ("The final choice: ")
                print ("You have two treasure boxes in front of you. Choose one of them.")
                treasure = int(input("Enter you choice - 1 or 2: "))
                treas = random.randint(1,2)
                if treasure == treas:
                    print ("Congratulations! You have won 1000 gold coins.")
                    user.gold += 1000
                    print ("You have won!")
                    user.show_info()
                else:
                    print ("You have chosen the wrong treasure box. You lost!")
                    print ("Game Over")
                    user.show_info()
            else:
                print ("End of Game!")
    else:
        print ("End of Game!")
        break


    
```
